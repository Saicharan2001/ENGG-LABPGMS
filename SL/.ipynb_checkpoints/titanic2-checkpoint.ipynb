```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df=pd.read_csv('train.csv')
titanic_df.head()
```

```python
titanic_df['Survived']=titanic_df['Survived'].map({0:'Died',1:'Survived'})
titanic_df.head()
```

```python
titanic_df.drop(['Parch','PassengerId','Name','SibSp','Ticket','Fare'],axis=1,inplace=True)
titanic_df.head()
```

```python
titanic_df['Embarked']=titanic_df['Embarked'].map({'S':'South','C':'Couth','Q':'Queens'})
titanic_df['Pclass']=titanic_df['Pclass'].map({1:'Luxury',2:'Economy',3:'Lower'})
titanic_df.head()
```

## DATA VISUALIZATIONS

```python
ax=sns.countplot(x='Pclass',hue='Survived',data=titanic_df,palette='Set1')
plt.show()
```

```python
interval=(0,18,35,60,120)
Cate=['Children','Teen','Adult','old']
titanic_df['New']=pd.cut(titanic_df.Age,interval,labels=Cate)
ax=sns.countplot(x='New',hue='Survived',data=titanic_df,palette='Set1')
plt.show()
```
